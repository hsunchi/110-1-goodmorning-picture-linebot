```python
import pandas as pd
```

```python
file_path = "./dataset.csv"
ds = pd.read_csv(file_path)
```

```python
ds
```

```python
def belong_category(content):

    if (
        content.find("新年") != -1
        or content.find("假日") != -1
        or content.find("中秋") != -1
        or content.find("虎") != -1
    ):
        return "holiday"
    elif content.find("早") != -1:
        return "morning"
    elif content.find("午") != -1:
        return "afternoon"
    elif content.find("聖誕") != -1 or content.find("平安") != -1:
        return "christmas"
    elif content.find("晚") != -1 or content.find("睡") != -1 or content.find("夢") != -1:
        return "night"
    else:
        return "others"
```

```python
ds["content"]
```

```python
ds["category"] = ds["content"].apply(belong_category)
```

```python
ds
```

```python
ds.to_csv("./dataset.csv")
```
